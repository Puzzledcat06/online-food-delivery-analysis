```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/cleaned_food_delivery.csv")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df['day_type'] = df['order_date'].dt.dayofweek.apply(
    lambda x: 'Weekend' if x >= 5 else 'Weekday'
)
```

```python
df["day_type"].value_counts()
```

```python
df['is_peak_hour'] = df['peak_hour'].apply(
    lambda x: 1 if str(x).lower() == 'yes' else 0
)
```

```python
df['is_peak_hour'].value_counts()
```

```python
df["profit_margin"].value_counts()
```

```python
df['profit_margin_pct'] = df['profit_margin'] * 100
```

```python
df[['profit_margin','profit_margin_pct']].head(10)
```

```python
def delivery_category(time):
    if time <= 45:
        return 'Fast'
    elif time <= 120:
        return 'Average'
    else:
        return 'Slow'

df['delivery_performance'] = df['delivery_time_min'].apply(delivery_category)
```

```python
df['delivery_performance'].value_counts(ascending=True)
```

```python
def age_group(age):
    if age < 18:
        return 'Below 18'
    elif age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 60:
        return '46-60'
    else:
        return '60+'

df['customer_age_group'] = df['customer_age'].apply(age_group)
```

```python
df['customer_age_group'].value_counts()
```
