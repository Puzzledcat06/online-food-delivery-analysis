```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/ONLINE_FOOD_DELIVERY_ANALYSIS.csv")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df=df.drop_duplicates()
```

```python
num_cols = [
    'order_value',
    'delivery_time_min',
    'distance_km',
    'customer_age'
]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = [
    'customer_gender',
    'payment_mode',
    'cuisine_type',
    'city',
    'area'
]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.loc[df['order_status'] == 'Cancelled', 'delivery_rating'] = np.nan
```

```python
df = df[df['order_value'] > 0]
df = df[df['delivery_time_min'] > 0]
df = df[df['distance_km'] > 0]
```

```python
df.to_csv("../data/processed/cleaned_food_delivery.csv", index=False)
```
