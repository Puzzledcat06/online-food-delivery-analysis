```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_food_delivery.csv")
```

```python
df.head()
```

```python
df['order_status'].value_counts().plot(kind="bar")
plt.title("Order Status Distribution")
plt.show()
```

```python
sns.histplot(df["order_value"],bins=40)
plt.title("Order Value Distribution")
plt.show()
```

```python
sns.histplot(df['delivery_time_min'], bins=40)
plt.title("Delivery Time Distribution")
plt.show()
```

```python
df['city'].value_counts().plot(kind='bar')
plt.title("City-wise Order Volume")
plt.xlabel("City")
plt.ylabel("Number of Orders")
plt.show()
```

```python
cuisine_counts = df['cuisine_type'].value_counts()

cuisine_counts.plot(kind='bar')
plt.title("Cuisine-wise Order Distribution")
plt.xlabel("Cuisine Type")
plt.ylabel("Number of Orders")
plt.show()
```

```python
payment_counts = df['payment_mode'].value_counts()

payment_counts.plot(kind='bar')
plt.title("Payment Mode Preferences")
plt.xlabel("Payment Mode")
plt.ylabel("Number of Orders")
plt.show()
```

```python
sns.scatterplot(
    x='distance_km',
    y='delivery_time_min',
    data=df,
    alpha=0.4
)
plt.title("Distance vs Delivery Time")
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Time (minutes)")
plt.show()
```

```python
sns.boxplot(
    x='delivery_rating',
    y='delivery_time_min',
    data=df
)
plt.title("Delivery Time vs Delivery Rating")
plt.xlabel("Delivery Rating")
plt.ylabel("Delivery Time (minutes)")
plt.show()
```

```python
cancelled_df = df[df['order_status'] == 'Cancelled']

cancel_reason_counts = cancelled_df['cancellation_reason'].value_counts()

cancel_reason_counts.plot(kind='bar')
plt.title("Cancellation Reason Distribution")
plt.xlabel("Cancellation Reason")
plt.ylabel("Number of Cancelled Orders")
plt.show()
```
